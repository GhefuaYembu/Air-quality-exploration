```python
#Rename the 'AQI2.5' to 'AQI2_5'
df.rename(columns={'AQI2.5': 'AQI2_5'}, inplace=True)
# Display the DataFrame to verify the changes
df
```

```python
#drop all rows with null values in AQI2_5
df = df.dropna(subset = ['AQI2_5'])
```

```python
df.shape
```

```python
#change the datatype of AQI2_5 from object to int to measure air quality
df['AQI2_5'] = pd.to_numeric(df['AQI2_5'], errors='coerce')
#df = df.dropna(subset=['AQI2_5'])
#df['AQI2_5'] = pd.to_numeric(df['AQI2_5'], errors='coerce').astype(int)
```

```python
df['AQI2_5'].dtypes
```

df.dtypes

```python
# Creating category column for AQI2.5
```

```python
#Create a new column with the different Categories of Air Quality
df.loc[df['AQI2_5']<=50, 'Category']= 'Good'
df.loc[(df['AQI2_5']>50) & (df['AQI2_5']<=100), 'Category']= 'Moderate'
df.loc[df['AQI2_5']>100, 'Category']= 'Bad'
```

```python
#Creating a month column
df['month_of_year'] = df['Date'].dt.strftime('%B')
```

```python
#extract only max values, visualize AQ category
df.head(-5)
```

```python
#veiwing the distribution of the Air quality index values
df.AQI2_5.plot(kind = 'hist')
```
